Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.

Answer:

Descriptive Statistics: Summarize and describe the main features of a dataset.
Example: In a class of 50 students, calculating the average marks, median, or standard deviation of their scores.

Inferential Statistics: Make predictions or inferences about a population based on a sample.
Example: Conducting a survey of 100 voters (sample) to predict election results for the entire country (population).

Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.

Answer:

Sampling: The process of selecting a subset (sample) from a population to make statistical inferences.

Random Sampling: Every individual has an equal chance of being selected.
Example: Selecting 50 students randomly from a university.

Stratified Sampling: Population is divided into groups (strata), and samples are taken proportionally from each group.
Example: Dividing students into male/female groups, then selecting randomly from each group.

Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.

Answer:

Mean: The average of all values.

Median: The middle value when data is arranged in order.

Mode: The most frequently occurring value.

Importance:

They summarize large datasets into a single representative value.

Used for decision-making, trend analysis, and understanding distributions.

Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?

Answer:

Skewness: Measures asymmetry of data distribution.

Positive Skew → Tail on the right (higher values are stretched).

Negative Skew → Tail on the left.

Kurtosis: Measures "tailedness" of the distribution.

High Kurtosis → More outliers (heavier tails).

Low Kurtosis → Fewer outliers (flatter distribution).

Positive Skew implies: Most values are lower, but a few very high values stretch the distribution.

Question 6: Compute the covariance and correlation coefficient between the following two datasets.
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Covariance matrix
cov_matrix = np.cov(list_x, list_y, bias=True)
covariance = cov_matrix[0, 1]

# Correlation coefficient
correlation = np.corrcoef(list_x, list_y)[0, 1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)


Output:

Covariance: 187.5
Correlation Coefficient: 0.9938586931957764

Question 7: Boxplot and outliers
import matplotlib.pyplot as plt
import seaborn as sns

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

sns.boxplot(data=data)
plt.title("Boxplot of Data with Outliers")
plt.show()


Explanation:

The box shows the interquartile range (IQR).

The line inside the box is the median.

The whiskers extend to values within 1.5 × IQR.

Any value beyond whiskers is an outlier.

Here, 35 is identified as an outlier because it lies well above the upper whisker.

Question 8: Relationship between advertising spend and sales
Explanation

Covariance: Shows the direction of relationship (positive = both increase together, negative = opposite).

Correlation: Standardized measure (between -1 and 1) showing strength and direction.

Correlation close to +1 → Strong positive relationship.

Correlation close to -1 → Strong negative relationship.

Correlation ≈ 0 → Weak or no linear relationship.

Python Code
import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Covariance
cov_matrix = np.cov(advertising_spend, daily_sales, bias=True)
covariance = cov_matrix[0, 1]

# Correlation
correlation = np.corrcoef(advertising_spend, daily_sales)[0, 1]

print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)


Output:

Covariance: 67500.0
Correlation Coefficient: 0.9976082467483912


✅ Very strong positive correlation, meaning more advertising spend → higher daily sales.

Question 9: Customer satisfaction survey analysis
Explanation

Summary Statistics to use:

Mean → Average satisfaction score.

Median → Middle score, less affected by outliers.

Standard Deviation → Spread of responses.

Visualization:

Histogram → Shows frequency distribution of satisfaction scores.

Boxplot → Can also be used to detect outliers.

Python Code
import matplotlib.pyplot as plt
import numpy as np

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Compute summary statistics
mean_score = np.mean(survey_scores)
std_dev = np.std(survey_scores)
median_score = np.median(survey_scores)

print("Mean:", mean_score)
print("Median:", median_score)
print("Standard Deviation:", std_dev)

# Plot histogram
plt.hist(survey_scores, bins=6, edgecolor='black')
plt.title("Histogram of Survey Scores")
plt.xlabel("Survey Score")
plt.ylabel("Frequency")
plt.show()


Output (printed values):

Mean: 7.4
Median: 7.0
Standard Deviation: 1.55


✅ Histogram will show most scores are around 7–9, meaning customers are generally satisfied.
